```python
# Install required libraries
!pip install geopandas osmnx matplotlib

import osmnx as ox
import matplotlib.pyplot as plt

# Define place of interest
place = "Ondo, Nigeria"

# Get river/stream features from OpenStreetMap
tags = {"waterway": ["river", "stream", "drain"]}

# Download the drainage network using the correct function
drainage = ox.features_from_place(place, tags)

#  Export to Shapefile
drainage.to_file("ondo_drainage.shp")

#  Export to GeoJSON (for web or GitHub visualization)
drainage.to_file("ondo_drainage.geojson", driver="GeoJSON")

# Plot drainage pattern
fig, ax = plt.subplots(figsize=(10, 10))
drainage.plot(ax=ax, color="blue", linewidth=1)
ax.set_title("Drainage Pattern of Ondo Town, Nigeria", fontsize=15)
plt.show()
```
