```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

# Simulate 4-band data (Blue, Green, Red, NIR)
np.random.seed(42)
blue = np.random.randint(0, 255, (50, 50))
green = np.random.randint(0, 255, (50, 50))
red = np.random.randint(0, 255, (50, 50))
nir = np.random.randint(0, 255, (50, 50))

# Stack into features
stacked = np.dstack((blue, green, red, nir))
X = stacked.reshape(-1, 4)

# Synthetic labels (3 land cover classes)
y = np.random.randint(0, 3, size=X.shape[0])

# Train Random Forest
clf = RandomForestClassifier(n_estimators=50, random_state=42)
clf.fit(X, y)

# Predict land cover
y_pred = clf.predict(X)
classified = y_pred.reshape(50, 50)

# Plot
plt.imshow(classified, cmap="tab20")
plt.title("Land Cover Classification (Demo)")
plt.colorbar()
plt.show()
```
